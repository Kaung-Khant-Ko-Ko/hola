# **Collecting Data**

```python
topics = ["인류의 진화", "미얀마의 역사", "물리학의 역사", "일본의 역사", "중국의 역사", "독일의 역사", "한국의 역사", "제1차 세계 대전", "제2차 세계 대전", "프랑스의 역사", "태국의 역사", "베트남의 역사", "인도네시아의 역사", "말레이시아의 역사", "필리핀의 역사", "캄보디아의 역사", "브루나이의 역사", "영국의 역사", "미국의 역사", "캐나다의 역사", "그리스의 역사", "로마의 역사", "헝가리의 역사", "오스트레일리아의 역사", "이탈리아의 역사", "스페인의 역사", "아르헨티나의 역사", "지구", "달", "행성", "우주", "컴퓨터", "인터넷", "생물학의 역사", "인공지능", "방탄소년단", "EXO", "외계 행성", "6.25 전쟁", "조선민주주의인민공화국", "대한민국", "미국"]
topics
```

```python
len(topics)
```

```python
para = ""
for topic in topics:
  result = wikipedia.search(topic)
  print(result)
  content_orig = wikipedia.page(result[0]).content
  content = content_orig.split("외부 링크")[0]
  content = content.split("참고 문헌")[0]
  content = content.split("각주")[0]
  st = re.sub(r"==.*?==+", "", content)
  st = re.sub("(#|\?|\\n|\\t)", "", st)
  para += st + "."
```

```python
data = para.split(".")
```

```python
len(data)
```

# **Saving the Extracted Sentences**

```python
data_df = pd.DataFrame(data)
data_df.head()
```
