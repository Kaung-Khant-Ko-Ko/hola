# **Collecting Data**

```python
topics = ["Evolución humana", "Historia de Birmania", "Historia de la física", "Historia de Japón", "Historia de China", "Historia de Alemania", "Historia de Corea", "Primera Guerra Mundial", "Segunda Guerra Mundial", "Historia de Francia", "Historia de Tailandia", "Historia de Vietnam", "Historia de Indonesia", "Historia de Malasia", "Historia de Filipinas", "Historia de Camboya", "Historia de Singapur", "Historia de Inglaterra", "Historia de América", "Historia de Canadá", "Historia de Roma", "Historia de Hungría", "Historia de África", "Historia de Italia", "Historia de la Argentina"]
topics
```

```python
len(topics)
```

```python
para = ""
for topic in topics:
  result = wikipedia.search(topic)
  print(result)
  content_orig = wikipedia.page(result[0]).content
  content = content_orig.split("Bibliografía")[0]
  content = content.split("Véase también")[0]
  content = content.split("Referencias")[0]
  st = re.sub(r"==.*?==+", "", content)
  st = re.sub("(#|/|\?|\\n|\\t|\[([0-9]*?)\]|\\u200b)", "", st)
  para += st + "."
```

```python
data = para.split(".")
```

```python
len(data)
```

# **Saving the Extracted Sentences**

```python
data_df = pd.DataFrame(data)
data_df.head()
```
