# **Collecting Data**

```python
topics = ["மனித பரிணாமம்", "படிவளர்ச்சிக் கொள்கை", "ஜெ. இராபர்ட் ஓப்பன்ஹீமர்", "மியான்மர்", "இயற்பியல்", "யப்பான்", "சீனா", "ஜெர்மனி", "கொரியா", "முதலாம் உலகப் போர்", "இரண்டாம் உலகப் போர்", "பிரான்சு", "தாய்லாந்து", "வியட்நாம்", "இந்தோனேசியா", "மலேசியா", "பிலிப்பீன்சு", "கம்போடியா", "புருனே-முரா மாவட்டம்", "சிங்கப்பூர்", "இங்கிலாந்து", "அமெரிக்க ஐக்கிய நாடுகள்", "கனடா", "கிரேக்கம் (நாடு)", "உரோம்", "அங்கேரி", "ஆத்திரேலியா", "எசுப்பானியா", "புவி", "அர்கெந்தீனா", "நிலா", "ஞாயிறு (விண்மீன்)", "கணினி", "பண்டைய எகிப்து", "நாட்சி ஜெர்மனி", "இட்லர்", "அலன் டூரிங்", "செயற்கை அறிவுத்திறன்", "அண்டம்", "இலங்கை", "வியட்நாம் போர்", "சூரியக் குடும்பம்", "இந்தியா", "இரால்ப் ஆழ்சர் ஆல்ப்பர்", "ஸ்டீவன் ஹாக்கிங்", "மேரி கியூரி", "ஐசாக் நியூட்டன்", "ஆல்பர்ட் ஐன்ஸ்டைன்", "அணுக்கரு ஆயுதங்கள்", "கலீலியோ கலிலி", "அலெக்சாண்டர் கிரகாம் பெல்", "அரிசுட்டாட்டில்", "கிராட்டனின் மிலோ"]
topics
```

```python
len(topics)
```

```python
para = ""
for topic in topics:
  result = wikipedia.search(topic)
  print(result)
  content_orig = wikipedia.page(result[0]).content
  content = content_orig.split("மேற்கோள்கள்")[0]
  st = re.sub(r"==.*?==+", "", content)
  st = re.sub("(#|\?|\\n|\\t|\[([0-9]*?)\]|\\u200d)", "", st)
  para += st + "."
```

```python
data = para.split(".")
```

```python
len(data)
```
